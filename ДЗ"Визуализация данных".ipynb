```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
d = pd.read_csv('2019.csv')
```

```python
d.describe()
```

```python
#Т10 самых щедрых стран
dft3 = d.sort_values(by=['Generosity'], ascending=False).head(10)
```

```python
y_pos = np.arange(len(dft3))
plt.bar(y_pos, dft3['Generosity'])
plt.xticks(y_pos, dft3['Overall rank'])
plt.xlabel('Overall rank')
plt.ylabel('Generosity')
plt.title('ТОП-10 щедрых стран')
#Можно сделать вывод, что не всегда, чем выше рейтинг счастья тем выше значение щедрости, на диаграмме видно что
#страны с невысоким рейтингом счастья имеют показатель щедрости выше среднего(0.1)
```

```python
plt.plot(d.Score, d['Social support'])
plt.title('Социальная поддержка')
plt.xlabel('Оценка')
plt.ylabel('Социальная поддержка') 
#График хорошо показывает, что чем выше показель социальной поддержки, тем выше уровень счастья.
```

```python
d.plot(kind = 'scatter', x = 'Score', y = "GDP per capita", color = "r")
plt.show()
#Диаграмма показывает, что чем выше показатель ВВП на душу населения тем выше уровень счастья
```

```python
d[['Perceptions of corruption', 'Freedom to make life choices' ]].plot(kind='box',
                                        title='')
#Можно сделать вывод, что чем выше показатель свободы делать жизненный выбор, 
#тем выше показатель восприятие коррупции.
```
