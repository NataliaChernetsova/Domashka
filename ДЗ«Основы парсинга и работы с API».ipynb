# Задание 1

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup 
import time
```

```python
URL = 'https://habr.com/ru/search/'
params = {
    'q': ['kfc', 'нетология'],
}
```

```python
res = requests.get(URL, params)
soup = BeautifulSoup(res.text, 'html.parser')
posts = soup.find_all('article', class_='post')
```

```python
habr_news = pd.DataFrame()
for art in posts:
    arts = art.find('a', class_='post__title_link').text
    titles = art.find('a', class_='post__title_link').text
    urls = art.find('a', class_='post__title_link').get('href')
    dates = art.find('span', class_='post__time').text
    texts = art.find('div', class_='post__text').text
    row = {'date': dates, 'title': titles, 'url': urls, 'text': texts}
    habr_news = pd.concat([habr_news, pd.DataFrame([row])])
habr_news
```

```python
email = ['666.26-06@mail.ru', 'chernetsovanton@gmail.com']
```

```python
url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
params = {'emailAddresses': email}
headers = {'Vaar-Version' : '0', 'Vaar-Header-App-Product' : 'hackcheck-web-avast'}
req = requests.post(url, data = json.dumps(params), headers = headers)
```

```python
dict_req = req.json()
df = pd.DataFrame()
```

```python
for k, v in dict_req.items():
    if k=='breaches':
        for p, d in dict_req['summary'].items():
            emails = p
            list_ = d['breaches']
            
            for i in list_:
                date = (dict_req['breaches'][str(i)]['publishDate'])
                source = (dict_req['breaches'][str(i)]['site'])
                description = (dict_req['breaches'][str(i)]['description'])
                
                row = {'Почта': emails, 'Дата учечки': date, 'Источник утечки': source, 'Описание учтечки': description}
                df = pd.concat([df, pd.DataFrame([row])])
```
