# Задание 1

```python
import pandas as pd
```

```python
df = pd.read_csv('ratings.csv')
```

```python
res = df.groupby('userId').agg({'rating': ['count'], 'timestamp': ['max', 'min']}).reset_index()
res.head()
```

```python
res['user lifetime'] = res['timestamp']['max'] - res['timestamp']['min']
```

```python
df2 = res[res['rating']['count'] > 100]
```

```python
result = df2[['userId', 'rating', 'user lifetime']]
result.head()
```
