```python
import pandas as pd
import numpy as np
```

```python
COL_N = ['surgery' , 'Age' , 'Hospital Number' , 'rectal temperature', 'pulse', 'respiratory_rate' , 
             'temperature_of_extremities', 'peripheral pulse', 'mucous membranes', 'capillary refill time', 
             'pain', 'peristalsis' , 'abdominal distension', 'nasogastric_tube', 'nasogastric reflux', 
             'nasogastric reflux PH', 'rectal examination - feces', 'abdomen' , 'packed cell volume', 
             'total_protein', 'abdominocentesis appearance', 'abdomcentesis total protein' , 'outcome', 
             'surgical lesion' , 'site of lesion', 'type, subtype', 'specific code', 'cp_data',]
```

```python
#Выводы: 1. Коликами либо болеют больше взрослые лошади, 
#либо молодые лошади не были учтены в данных по какой-то другой причине
#2.В большинстве своем(75%) болезнь проходит без операции, но в 50% случаев необходима операция  
#3.В 75% случаев болезнь проходит при слегка повышенной температуре (38.5) 
#при максимальной температуре 40.8 лошадь может умереть так как другие показатели ухудшаются.
#4.При ухудшении состояния лошади пульс значительно повышается, что тоже может привести к летальному исходу
#5. Можно сделать вывод, что если у лошади повышенная температура и учащённый пульс, 
#а также если вздутие живота тяжелое и у лошади снижается активность в кишечнике, 
#то мы имеем дело с тяжелой формой заболевания.
```

# Выбросы

```python
q1 = df1.Age.quantile(0.25)
q3 = df1.Age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df1[df1.Age.between(lower_bound, upper_bound, inclusive=True)]
```

```python
pd.concat([df1, remove_outliers]).drop_duplicates(keep=False)
# ошибка ввода данных, так как значение может быть 1 = Взрослая лошадь либо,2 = Молодой (<6 месяцев) 
#оставим как есть
```

```python
q1 = df1['pulse'].quantile(0.25)
q3 = df1['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers_t = df1[df1['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('abdominal distension')
pd.concat([df1, remove_outliers_t]).drop_duplicates(keep=False)
#remove_outliers_t
#Вывод:лошади с пульсом больше 150 умерли либо их усыпили
```

# пропуски

```python
for col in df.columns:
    pct_missing = df[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
for col in df1.columns:
    pct_missing = df1[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
#Пропуски в rectal temperature заменим на среднее значение, так как болезнь в большинстве
#случаев проходит со средней температурой

df1['rectal temperature'].fillna(df1['rectal temperature'].mean(), inplace=True)
df1['rectal temperature'].fillna(df1['rectal temperature'].mean()).isna().sum()
```

```python
#В столбце Pulse и outcome пропусков мало, поэтому на анализ данных пропуски не повлияют, оставим как есть
```

```python
#abdominal distension это важный качественный показатель поэтому заполним пропуски его  модой
df1['abdominal distension'].fillna(df1['abdominal distension'].mode()[0], inplace=True)
```

```python
#nasogastric_tube заполняем с группировкой по pulse
df1.nasogastric_tube.fillna(df1.groupby('pulse').nasogastric_tube.transform('median'), inplace=True)
```

```python
res = df1.fillna(df1.groupby('Age').pulse.transform('mean'), inplace=True)
res = df1.groupby('Age').pulse.describe()
res
```
